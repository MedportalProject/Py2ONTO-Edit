# Initalization of Py2ONTO-Edit Tools
## load HumanDO.owl

```python
# load HumanDO.owl
humanDO = EDIT_ONTO("./HumanDO.owl")
```

## Step 1: Segmentation Ontology

#### 1.1 Segmentation method 1: Global extraction method
#### Get all data under a single root node and store to new_onto.owl

```python
humanDO.cut_part_onto('orofacial cleft')
```

#### 1.2 Segmentation method 2: Selective depth extraction method

```python
humanDO.cut_part_onto_selection('bacterial infectious disease', 'commensal bacterial infectious disease', 'bartonellosis','opportunistic bacterial infectious disease', 'anthrax disease')
```

## Step 2: Terms Translation

#### 2.1 Export all class data from ontology into csv file

```python
humanDO.owl_to_csv("./new_onto.owl")
```

#### 2.2 Terms Translation

#### 2.2.1 Translation with Argos-translate Python package

```python
# translate ENGLISH to GERMAN 
# humanDO.translate_terms_with_Package("./part_onto.csv", "en2de")
```

```python
# translate ENGLISH to CHINESE
humanDO.translate_terms_with_Package("./part_onto.csv", "en2zh")
```

#### 2.2.2 Translation with DeepL
#### translate_terms_with_deepl('./part_onto.csv', 'translation_mode', 'deepL_API_Key')

> translation_mode:  
> en2zh:English to Chinese  
> en2fr:English to French  
> en2de: English to German  

```python
# translate ENGLISH to FRENCH via DeepL
humanDO.translate_terms_with_deepl("./part_onto.csv","en2fr",'your_deepL_API_Key')
```

```python
# translate ENGLISH to CHINESE via DeepL
humanDO.translate_terms_with_deepl("./part_onto.csv","en2cn",'your_deepL_API_Key')
```

#### 2.2.3 Translation with ChatGLM-130B
#### translate_terms_with_GLM('./part_onto.csv','api_key', 'translation_mode', 'glm-4')

```python
# translate ENGLISH to German via ChatGLM-130B
humanDO.translate_terms_with_GLM('./part_onto.csv', 'your_glm_API_key', 'en2de', 'glm-4')
```

```python
# translate ENGLISH to Chinese via ChatGLM-130B
humanDO.translate_terms_with_GLM('./part_onto.csv', 'your_glm_API_key', 'en2zh', 'glm-4')
```

#### 2.2.4 Translation with Gemini
#### translate_terms_with_gemini('./part_onto.csv','api_key', 'translation_mode', 'gemini-pro')

```python
# translate ENGLISH to German via Gemini
humanDO.translate_terms_with_gemini('./part_onto.csv','your_gemini_API_Key', 'en2de','gemini-pro')
```

```python
# translate ENGLISH to Chinese via Gemini
humanDO.translate_terms_with_gemini('./part_onto.csv','your_gemini_API_Key', 'en2zh','gemini-pro')
```
